````python
"""
# write a program to find the grades based on student marks
1. if marks in bewteen 80 to 100 - A Grade
2. if marks in bewteen 60 to 79 - B Grade
3. if marks in bewteen 40 to 59 - C Grade
4. if marks in bewteen 25 to 39 - D Grade
5. if marks in bewteen 0 to 24 -  Fail
6. if Other then above marks output is - please enter correct marks
"""
marks = int(input("enter student marks here:"))
if (marks>=80 and marks<=100):
    print(marks, "A Grade")
elif (marks>=60 and marks<=79):
    print(marks, "B Grade")
elif (marks>=40 and marks<=59):
    print(marks, "C Grade")
elif (marks>=25 and marks<=39):
    print(marks, "D Grade")
elif (marks>=0 and marks<=24):
    print(marks, "Fail")
else:
    print("please enter correct marks")


````
